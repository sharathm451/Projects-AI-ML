```python
#PART II: Draw ACF and PACF Plot Charts in Python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

prices=pd.read_csv("PriceByDays.csv",parse_dates=True,index_col=0)
sm.graphics.tsa.plot_acf(prices.values.squeeze(), lags=15)
sm.graphics.tsa.plot_pacf(prices.values.squeeze(), lags=15)
plt.show()
```
